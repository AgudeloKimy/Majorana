```python
#################################################################################
delta=2
def condition_model(l):
  d={} # Se inicializa un dicionario vacío
  ll=set(l)
  for n in ll:
    if l.count(n)>1: # Mayor que 1 porque si se repite más veces quiere decir que hay más 1 n.
      d['nu']=n # Entonces m=n
  if d.get('nu'): #.get--> forma segura de pedir un elemnto aunque no exista. Sí existe m volver a contar
    ll.remove(d['nu']) # removemos el elemento sí existe y ese el elemento que necesitamos como nu.
    #if len(ll)>0:
      #d['nu']=list(ll)[0] #nu es el número que se repite, por lo tanto se toma el primer elemento
    d['s']= 2*d['nu']/delta  # Ecuación del modelo.
    return d
  else:
    return {}
```

```python
solutions_interest['condition_model']=solutions_interest['repeated'].apply(condition_model)
```

```python
def Massless(x,s):
  if True:
    xx=x.copy()
    d= []
    if len(x)>0:
      for x in itertools.permutations(x,2):
        if x[0]<=x[1]:
          d.append(x)
    for j in d:
      if abs(sum(j))== abs(s):
        d.remove(j)
        j=list(j)
        for elements in j:
          if elements in xx:
            xx.remove(elements)

  return xx
```
